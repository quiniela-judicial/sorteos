```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def remove_s(ele):
    if ele.endswith('S'):
        return ele[:-1]
    else:
        return ele
```

```python
df = pd.read_csv("../data/implicados.csv")
df['Tipo implicado'] = df['Tipo implicado'].apply(remove_s)
df.head()
```

```python
df[df['Nombre implicado'].str.contains('DE VIDO')]
```

```python
df[df['Nombre implicado'] == 'DE VIDO JULIO MIGUEL']['Tipo implicado'].value_counts().plot(kind='bar')
plt.show()
```
