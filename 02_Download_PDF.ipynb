```python
import os

import requests
import pandas as pd
```

```python
base_dir = '../data/'
```

```python
def create_file_name(ele):
    """
    
    """
    rest, text = ele.split('&id=')
    file_id, rest = text.split('&', 1)
    
    return file_id + '.pdf'


def download_file(url, file_path):
    """
    Download the PDF file
    """
    # NOTE the stream=True parameter
    r = requests.get(url, stream=True)
    with open(file_path, 'wb') as f:
        for chunk in r.iter_content(chunk_size=1024): 
            if chunk: # filter out keep-alive new chunks
                f.write(chunk)
```

```python
df_resolutions = pd.read_csv(base_dir + 'resolucion.csv')
df_resolutions.head()
```

```python
# Create PDF name
files_names = df_resolutions['PDF link'].apply(create_file_name)
df_resolutions['PDF nombre'] = files_names

df_resolutions.to_csv(base_dir + 'resolucion.csv', index=False)
```

```python
# Cantidad de PDF a bajar
amount_download = df_resolutions.shape[0]

# PDF ya bajados
ready_files = os.listdir(base_dir + 'PDFs')

for index, row in df_resolutions.iterrows():
    # Progress text
    progress_text = '\rDescargando PDF numero {0} de {1}'
    print(progress_text.format(index + 1, amount_download), end='')
    
    # Si el pdf ya esta bajado, no lo vuelvo a bajar
    if row['PDF nombre'] in ready_files:
        continue
    
    # Download PDF
    path = base_dir + 'PDFs/' + row['PDF nombre']
    download_file(row['PDF link'], path)
```
