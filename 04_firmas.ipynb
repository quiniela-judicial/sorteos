```python
import os

import pandas as pd
```

```python
base_dir = '../data/'
```

```python
df_resolutions = pd.read_csv(base_dir + 'resolucion.csv')
df_resolutions
```

```python
rows = []

for index, row in df_resolutions.iterrows():
    # No se por que, pero un pdf esta vacio. Lo ignoro
    path = base_dir + 'PDFs/' + row['PDF nombre']
    if os.stat(path).st_size == 0:
        continue
        
    path = path.replace('.pdf', '.txt')

    with open(path) as file:
        firms = list(set([line for line in file.readlines() 
                          if line.startswith('Firmado')]))

        # Split in :
        firms = [firm.split(':')[1] for firm in firms]
        # Get the person name and the job
        names = [firm.split(',')[0] for firm in firms]
        jobs = [firm.split(',')[1] if ',' in firm else '' for firm in firms]
        
        # Remove whites spaces
        names = [' '.join(name.split()).upper() for name in names]
        jobs = [' '.join(job.split()).upper() for job in jobs]
        
        # Some replaces
        jobs = [job.replace('CAMARA', 'CÁMARA').replace('SECRETARIA', 'SECRETARIO').replace('JUEZA', 'JUEZ') 
                for job in jobs]
        
        for name, job in zip(names, jobs):
            row_partial = row.append(pd.Series([name, job], index=['Firma', 'Cargo']))
            rows.append(row_partial.tolist())
```

```python
df_resolutions = pd.DataFrame(rows, columns=row_partial.index)
df_resolutions.head()
```

```python
df_resolutions.to_csv('../data/resolucion.csv', index=False)
```
